# 2. Implementar el algoritmo White Patch para librarnos de las diferencias de color de iluminación.

```python
import numpy as np

def white_patch(img):
    # Convertir la imagen a float32
    img = img.astype(np.float32)

    # Encontrar el valor máximo en cada canal de color
    max_color = np.max(img, axis=(0, 1))

    # Dividir cada canal de color por su valor máximo
    img = img / max_color

    return img
```

```python
import cv2

image_path = './white_patch/test_blue.png'  
img = cv2.imread(image_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

corrected_img = white_patch(img)
```

```python
# Mostrar la imagen de las coordenadas cromáticas
import matplotlib.pyplot as plt

# Normalizar los valores al rango 0-255 para mostrar la imagen
normalized_img = (corrected_img * 255).astype(np.uint8)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()

plt.imshow(cv2.cvtColor(normalized_img, cv2.COLOR_BGR2RGB))
plt.show()
```
