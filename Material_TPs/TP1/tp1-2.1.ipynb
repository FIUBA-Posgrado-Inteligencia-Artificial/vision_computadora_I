```python
import matplotlib.pyplot as plt
import cv2

image_path = './img1_tp.png' 
img_gray = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

plt.imshow(img_gray, cmap='gray')
plt.show()
```

```python
import matplotlib.pyplot as plt
import cv2

image_path2 = './img2_tp.png' 
img_gray2 = cv2.imread(image_path2, cv2.IMREAD_GRAYSCALE)

plt.imshow(img_gray2, cmap='gray')
plt.show()
```

```python
num_bins = 256

plt.hist(img_gray.ravel(), bins=num_bins, color='gray')
plt.title('Histograma de la primera imagen en escala de grises')
plt.show()

plt.hist(img_gray2.ravel(), bins=num_bins, color='gray')
plt.title('Histograma de la segunda imagen en escala de grises')
plt.show()
```

Los histogramas pueden ser útiles como características en modelos de clasificación/detección de imágenes porque proporcionan información sobre la distribución de intensidades de píxeles en la imagen. Sin embargo, los histogramas no conservan la información espacial de la imagen (es decir, dónde se encuentran los píxeles en la imagen), por lo que pueden no ser suficientes por sí solos para algunas tareas de clasificación/detección de imágenes. A menudo se utilizan en combinación con otras características.
