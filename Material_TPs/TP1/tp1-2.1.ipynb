```python
import matplotlib.pyplot as plt
import cv2

image_path = './img1_tp.png' 
img_gray = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

plt.imshow(img_gray, cmap='gray')
plt.show()
```

```python
import matplotlib.pyplot as plt
import cv2

image_path2 = './img2_tp.png' 
img_gray2 = cv2.imread(image_path2, cv2.IMREAD_GRAYSCALE)

plt.imshow(img_gray2, cmap='gray')
plt.show()
```

```python
num_bins = 256

plt.hist(img_gray.ravel(), bins=num_bins, color='gray')
plt.title('Histograma de la primera imagen en escala de grises')
plt.show()

plt.hist(img_gray2.ravel(), bins=num_bins, color='gray')
plt.title('Histograma de la segunda imagen en escala de grises')
plt.show()
```

Los histogramas pueden ser tiles como caractersticas en modelos de clasificacin/deteccin de imgenes porque proporcionan informacin sobre la distribucin de intensidades de pxeles en la imagen. Sin embargo, los histogramas no conservan la informacin espacial de la imagen (es decir, dnde se encuentran los pxeles en la imagen), por lo que pueden no ser suficientes por s solos para algunas tareas de clasificacin/deteccin de imgenes. A menudo se utilizan en combinacin con otras caractersticas.
